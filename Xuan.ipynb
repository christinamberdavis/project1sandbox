```python
# Set up Dependencies and Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import dataframe_image as dfi
```

```python
# Load the CSV into a Pandas DataFrame
raw_data_df = pd.read_csv("Resources_xuan/User0_credit_card_transactions.csv", na_values=' ')

# Display sample data
raw_data_df.head()
```

```python
# Select the columns needed for analysis and created a sub data set.
transaction_type_raw_df = raw_data_df[['Year', 'Amount', 'Use Chip', 'Errors?', 'Is Fraud?' ]]
# Fill in the empty cell with N/A
transaction_type_df = transaction_type_raw_df.fillna('N/A')
# Export the transaction_type_df into a csv
transaction_type_df.to_csv("Resources_xuan/data_xuan.csv")

transaction_type_df.head()
```

## "The Proportion of The Three Transaction Methods from 2002 to 2020"

```python
# Create a pie chart display the portion of each transation type from 2002 to 2020
# Calculate the total transactions of each transaction type
num_swipe_trans = 0
num_online_trans = 0
num_chip_trans = 0

for i in range(1,transaction_type_df.index.size):
    if transaction_type_df['Use Chip'][i] == 'Swipe Transaction':
        num_swipe_trans = num_swipe_trans + 1
    if transaction_type_df['Use Chip'][i] == 'Online Transaction':
        num_online_trans = num_online_trans + 1
    if transaction_type_df['Use Chip'][i] == 'Chip Transaction':
        num_chip_trans = num_chip_trans + 1
# Labels for the sections of our pie chart
labels = ["Swipe", "Online", "Chip"]

# The values of each section of the pie chart
sizes = [num_swipe_trans, num_online_trans, num_chip_trans]

# The colors of each section of the pie chart
colors = ["lightgreen", "lightcoral", "lightskyblue"]

# Plot the pie chart
#colors
colors = ['#66b3ff', '#ff9999','#99ff99']

plt.title("The Proportion of The Three Transaction Methods from 2002 to 2020")
plt.pie(sizes, labels=labels, textprops={"fontsize":12}, colors=colors, wedgeprops=
       {'edgecolor':'white'}, startangle=90,
        autopct="%1.1f%%", shadow=True)


# Save the figure
plt.savefig("output_data_xuan/Fig1.png")
# Show plot
plt.show()
```

## Total transaction volume (in each transaction type) each year between 2002 to 2020.

```python
# Put years into a list for for loop (and later for plot labels)
Years = [*set(transaction_type_df['Year'])]

# Create empty lists to fill with number of transactions each year for each transaction type (for plotting)
swipe_num_by_year = []
online_num_by_year = []
chip_num_by_year = []

#Calculate the nunber of transations each year for each transaction type
for year in Years:
    num_swipe_single_year = len(transaction_type_df[(transaction_type_df['Year'] == year) & (transaction_type_df['Use Chip'] =='Swipe Transaction')])
    num_online_single_year = len(transaction_type_df[(transaction_type_df['Year'] == year) & (transaction_type_df['Use Chip'] =='Online Transaction')])
    num_chip_single_year = len(transaction_type_df[(transaction_type_df['Year'] == year) & (transaction_type_df['Use Chip'] =='Chip Transaction')])
    swipe_num_by_year.append(num_swipe_single_year)
    online_num_by_year.append(num_online_single_year)
    chip_num_by_year.append(num_chip_single_year)

transaction_types = ['Swipe Transaction','Online Transaction','Chip Transaction']

plot_2_df = pd.DataFrame(columns=Years, index= transaction_types)
plot_2_df.loc['Swipe Transaction']= swipe_num_by_year
plot_2_df.loc['Online Transaction']= online_num_by_year
plot_2_df.loc['Chip Transaction']= chip_num_by_year
plot_2_df.head()
```

### Create a line chart showing transaction volume in each year.

```python
ax = plot_2_df.T.plot(figsize=(14, 6))
ax.set_ylabel('Total Transaction Volum', fontsize=12)
ax.set_xlabel('Years', fontsize=12)
ax.set_title('Total transaction volume each year between 2002 to 2020', fontsize=12)
ax.set_xticks(plot_2_df.columns)
plt.savefig("output_data_xuan/Fig3.png")
plt.show()
```

### Create a dataframe and save in png showing transaction volume in each year.

```python
plot_2_df['Total'] = [num_swipe_trans, num_online_trans, num_chip_trans]
plot_2_df.style.set_precision(10)
plot_2_df.head()
```

```python
dfi.export(plot_2_df, "output_data_xuan/dataframe.png")
```

## Error rate and fraud rate among three transaction types between 2002 to 2020.

```python
# Create empty lists (for plotting)
num_total_list =[]
num_error_list =[]
num_fraud_list = []
error_rate_list = []
fraud_rate_list = []

# Calculate the error rate in each transaction types and prep for plotting
for a in transaction_types:
    a_error = len(transaction_type_df[(transaction_type_df['Use Chip'] == a) & (transaction_type_df['Errors?']!='N/A')])
    a_total = len(transaction_type_df[transaction_type_df['Use Chip'] == a])
    a_fraud =  len(transaction_type_df[(transaction_type_df['Use Chip'] == a) & (transaction_type_df['Is Fraud?']=='Yes')])
    #error_rate = int(a_error)/int(a_total)
    #fraud_rate = int(a_fraud)/int(a_total)
    error_rate = "{:.2%}".format(int(a_error)/int(a_total))
    fraud_rate = "{:.2%}".format(int(a_fraud)/int(a_total))
    num_error_list.append(a_error)
    num_fraud_list.append(a_fraud)
    num_total_list.append(a_total)
    error_rate_list.append(error_rate)
    fraud_rate_list.append(fraud_rate)

#error_rate = "{:.2%}".format(int(a_error)/int(a_total))
#fraud_rate = "{:.2%}".format(int(a_fraud)/int(a_total))
plot_df_1 = pd.DataFrame({"Total Errors":num_error_list, "Error Rate": error_rate_list, "Total Fraud": num_fraud_list, "Fraud Rate": fraud_rate_list, "Total Transactions": num_total_list}, index = transaction_types)
plot_df_1.head()


```
